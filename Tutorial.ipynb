### Latitude Dependent Scale Bar

This notebook provides usage examples for the scalebar library.  The library allows for the generation of latitude dependent scale bars for the following projections:

* Mercator
* Lambert Conformal Conic
* Polar Stereographic

Scalebars can be generated with input from either a raster data product or a projection string and lat/lon extent parameter.

```python
import os #Imported only to cleanup.
from IPython.display import SVG  #For visualization of the rendered SVG

import scalebar
```

### Mercator from image
Access to the ScaleBar class is primarliy made through two class methods (constructors).  These are:

* `from_image` in the case where the input is an image
* `from_projstring` in the case where the input is a projection string

The input, either image or projection string must be in a format that is supported by GDAL.

```python
img = scalebar.examples.get_path('Mars_MGS_MOLA_ClrShade_MAP2_0.0N0.0_MERC.tif')
s = scalebar.bar.ScaleBar.from_image(img, outputname='marsmerc.svg')
SVG(filename='marsmerc.svg')
```

The `s` object has attributes to help track the provenence of the scalebar.  We could add a `__repr__` that made printing that metadata easy.

```python
print s.spatialreference
print
print s.latlon_bounds
print
print s.name
print
```

### Marcetor from a projection string

```python
wkt = """PROJCS["Mercator_MARS",
GEOGCS["GCS_MARS",
DATUM["MARS",
SPHEROID["MARS",3396190,169.8944472236118]],
PRIMEM["Reference_Meridian",0],
UNIT["Degree",0.017453292519943295]],
PROJECTION["Mercator_1SP"],
PARAMETER["central_meridian",0],
PARAMETER["false_easting",0],
PARAMETER["false_northing",0],
UNIT["Meter",1],
PARAMETER["latitude_of_origin",0.0]]"""
s = scalebar.bar.ScaleBar.from_projstring(wkt, (0,0,180,65), outputname='marsmerc_wkt.svg')
SVG(filename='marsmerc_wkt.svg')
```

###Lambert from Image

```python
img = scalebar.examples.get_path('Lunar_LRO_LOLA_Shade_MAP2_90.0N20.0_LAMB.tif')
s = scalebar.bar.ScaleBar.from_image(img, outputname='lunarlamb.svg')
SVG(filename='lunarlamb.svg')
```

###Lambert from WKT string
This projection string is pulled from spatialreference.org.  The standard parallels have been modified from -20, 20 as proj4 fails when $\theta_{1} = -\theta_{2}$

```python
wkt = """PROJCS["Moon_Lambert_Conformal_Conic_AUTO",
        GEOGCS["Moon 2000",
            DATUM["D_Moon_2000",
            SPHEROID["Moon_2000_IAU_IAG",1737400.0,0.0]],
            PRIMEM["Greenwich",0],
            UNIT["Decimal_Degree",0.0174532925199433]],
            PROJECTION["Lambert_Conformal_Conic"],
            PARAMETER["False_Easting",0],
            PARAMETER["False_Northing",0],
            PARAMETER["Central_Meridian",0],
            PARAMETER["Standard_Parallel_1",43],
            PARAMETER["Standard_Parallel_2",73],
            PARAMETER["Latitude_Of_Origin",0],
            UNIT["Meter",1]]"""
s = scalebar.bar.ScaleBar.from_projstring(wkt, (33.25, 36, 60, 78.11), outputname='lunarlamb_wkt')
SVG(filename='lunarlamb_wkt')
```

### Polar Stereographic from Image

```python
img = scalebar.examples.get_path('Mars_MGS_MOLA_ClrShade_MAP2_90.0N0.0_POLA.tif')
s = scalebar.bar.ScaleBar.from_image(img, outputname='marspolar.svg')
SVG(filename='marspolar.svg')
```

### Polar Stereographic from WKT
This is a south pole scalebar, so we must alter the `cliplat` parameter.  This keyword defines where the latitude is clipped out of the scalebar.  The default is the equator.

```python
wkt = """PROJCS["Mars_South_Pole_Stereographic",
        GEOGCS["Mars 2000",
            DATUM["D_Mars_2000",
            SPHEROID["Mars_2000_IAU_IAG",3396190.0,169.89444722361179]],
            PRIMEM["Greenwich",0],
            UNIT["Decimal_Degree",0.0174532925199433]],
            PROJECTION["Stereographic"],
            PARAMETER["False_Easting",0],
            PARAMETER["False_Northing",0],
            PARAMETER["Central_Meridian",0],
            PARAMETER["Scale_Factor",1],
            PARAMETER["Latitude_Of_Origin",-90],
            UNIT["Meter",1]]"""
name = 'marspolar_wkt.svg'
s = scalebar.bar.ScaleBar.from_projstring(wkt, (0, -90, 180,-40), outputname=name, cliplat=-90)
SVG(filename=name)
```

### Mercator spanning equator 
'It's a song...'

```python
wkt = """PROJCS["Mercator_Mars2000_Sphere",
         GEOGCS["Mars_2000_Sphere",
             DATUM["D_Mars_2000",
             SPHEROID["Mars_2000",3396190,0]],
             PRIMEM["Reference_Meridian",0],
             UNIT["degree",0.0174532925199433]],
             PROJECTION["Mercator_1SP"],
             PARAMETER["central_meridian",0],
             PARAMETER["scale_factor",1],
             PARAMETER["false_easting",0],
             PARAMETER["false_northing",0],
             UNIT["meter",1,
             AUTHORITY["EPSG","9001"]]
         ]"""
name = 'mars_merc_cross_equator.svg'
#extent -5S to 15N (lat) and -15W to 15E
s = scalebar.bar.ScaleBar.from_projstring(wkt, (-5, -15, 15, 15), outputname=name,
                                          lon_minor_ticks=[12.5],
                                          lon_major_ticks=[25, 50, 75, 100],
                                          cliplat=-5.0)
SVG(filename=name)
```
